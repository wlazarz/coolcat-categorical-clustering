### Muschrooms Dataset
Mushroom [Dataset]. (1981). UCI Machine Learning Repository. doi.org/10.24432/C5959T.

```python
import pandas as pd
from coolcat import Coolcat

df = pd.read_csv('agaricus-lepiota.csv', sep=';')
df_class = df['class']
df.drop('class', axis=1, inplace=True)
for c in df.columns:
    df[c] = df[c].fillna(df[c].mode()[0])

df.head()
```

```python
x=Coolcat(21)
labels = x.fit_predict(df)
```

```python
df['class'] = df_class
df['labels'] = labels

result = df.groupby(['labels', 'class']).size().unstack(fill_value=0)
result = result[['p', 'e']] if 'p' in result.columns and 'e' in result.columns else result
```
